# Una breve aclaración sobre la IA

Imagina que tienes un robot de juguete, pero no puede hacer nada a menos que lo muevas. La Inteligencia Artificial (IA) es como darle a ese robot un cerebro diminuto para que aprenda a hacer cosas por sí solo.

Ese cerebro diminuto **NO** es otra cosa que una versión muy muy avanzada de esta fórmula matemática:

$$Y = \alpha + \beta + \epsilon$$


Es simple español, sería algo como:

$$Resultado = Pesos * Inputs$$

Donde:

Resultado = Es el resultado esperado por la persona que usa la IA

Pesos = Son valores pre-estimados, los cualaes se usan para los cálculos con los números dados por del usuario

Inputs = Los números que tu como usuario colocas.

Por ejemplo:

```python
Pesos = 0.5
Inputs = 10 # Número de entradas que el usuario coloca

Resultado = Pesos * Inputs

print(f"El resultado de este cálculo es: {Resultado}")
```

Otro ejemplo:

```python
Pesos = 0.5
Inputs = 20 # Número de entradas que el usuario coloca

Resultado = Pesos * Inputs

print(f"El resultado de este cálculo es: {Resultado}")
```

Como notarán esto es solo una operación matemática. En la IA es igual pero se usan palabras.... o a lo mejor no...

# Cómo el System Prompt y la Temperatura cambian los resultados de la IA

Hay dos "palancas" principales que controlan cómo responde una IA:

### 1. System Prompt (Mensaje de Sistema)
Es una instrucción invisible para el usuario final que le dice a la IA **cómo comportarse**. Por ejemplo:
- Si le dices _"Eres un asistente bíblico que solo reporta lo que dice el texto"_, la IA se limitará a citar pasajes.
- Si le dices _"Eres un poeta"_, la misma pregunta recibirá una respuesta completamente diferente.

**Cambia el system prompt en la interfaz de abajo y observa cómo cambian las respuestas.**

### 2. Temperatura
Es un número entre **0** y **1** que controla qué tan "creativa" o "aleatoria" es la respuesta:
- **Temperatura = 0**: La IA siempre elige la palabra más probable. Las respuestas son consistentes y repetibles.
- **Temperatura = 1**: La IA explora palabras menos probables. Las respuestas son más variadas y creativas, pero pueden ser menos precisas.

**Prueba hacer la misma pregunta con temperatura 0 y luego con temperatura 1 para ver la diferencia.**

# Conclusiones

## 1. La IA no es magia, es matemática
Como vimos al inicio, detrás de toda IA hay una operación fundamental: **Resultado = Pesos × Inputs**. No importa cuán sofisticada sea la respuesta que veas en pantalla, todo se reduce a multiplicaciones entre números. La IA no "piensa", no "cree", no "entiende". Calcula.

## 2. Las computadoras no leen palabras, leen números
Los embeddings nos mostraron que un versículo como Gálatas 5:14 — *"Amarás a tu prójimo como a ti mismo"* — es convertido en una lista enorme de números decimales antes de que la IA pueda hacer cualquier cosa con él. Esa es la verdadera "lectura" de la máquina.

## 3. El System Prompt define el carácter de la IA
El mismo modelo de IA, con la misma pregunta, produce respuestas radicalmente diferentes dependiendo del mensaje de sistema:

- Con un **prompt neutral** (`mensaje_de_systema`): la IA reporta lo que dice el texto bíblico, muestra diferencias entre pasajes, y no toma partido.
- Con un **prompt sesgado** (`mensaje_de_systema_extremista`): la IA presenta una sola interpretación como verdad absoluta, omite pasajes incómodos, armoniza contradicciones a la fuerza, y trata con condescendencia a quien cuestione.

**La IA no tiene opinión propia.** Adopta la postura que le dictes. Eso la convierte en una herramienta poderosa, pero también en un arma peligrosa si el prompt es diseñado con sesgo — especialmente cuando ese sesgo es sutil y suena "educado".

## 4. La temperatura controla la variabilidad
- **Temperatura baja (0)**: respuestas consistentes, predecibles y deterministas.
- **Temperatura alta (1)**: respuestas más variadas y creativas, pero potencialmente menos precisas.

Esto significa que dos personas pueden hacer la misma pregunta al mismo modelo y recibir respuestas diferentes simplemente porque la temperatura era distinta.

## 5. La lección más importante: pensamiento crítico
La IA no reemplaza el criterio humano. Puede presentar información falsa con total confianza, omitir datos importantes, o mezclar hechos con interpretaciones de forma imperceptible — **todo dependiendo de quién escribió el prompt**.

Antes de confiar en cualquier respuesta generada por IA, pregúntate:
- **¿Quién diseñó las instrucciones de esta IA?**
- **¿Qué sesgo podría tener el prompt que no estoy viendo?**
- **¿La IA está mostrando el texto completo, o solo lo que conviene?**

La IA es un espejo del prompt. Si el prompt es honesto, el espejo refleja con claridad. Si el prompt es sesgado, el espejo distorsiona — y lo hace con una elocuencia que puede engañar a cualquiera.
